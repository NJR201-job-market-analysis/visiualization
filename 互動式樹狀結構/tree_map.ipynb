```python
# 列表-下拉式選單

import tkinter as tk
from tkinter import ttk


def insert_tree_items(parent, tree, items):
    for key, value in items.items():
        node = tree.insert(parent, 'end', text=key)
        if isinstance(value, dict):
            insert_tree_items(node, tree, value)
        elif isinstance(value, list):
            for item in value:
                tree.insert(node, 'end', text=item)

def create_tree_view():
    root = tk.Tk()
    root.title("互動式人際網絡")

    tree = ttk.Treeview(root)
    tree.pack(expand=True, fill='both')

    # 插入主節點
    for main_key, main_value in topic_tree.items():
        main_node = tree.insert('', 'end', text=main_key)
        if isinstance(main_value, dict):
            insert_tree_items(main_node, tree, main_value)

    # 展開所有節點
    for item in tree.get_children():
        tree.item(item, open=True)

    root.mainloop()

create_tree_view()
```

```python
import plotly.express as px
import pandas as pd

# 假設 topic_tree 已經定義
# topic_tree = {...}  # 這裡不重新定義 topic_tree

def create_tree_data(parent, items):
    data = []
    for key, value in items.items():
        data.append({"parent": parent, "label": key})
        if isinstance(value, dict):
            data.extend(create_tree_data(key, value))
        elif isinstance(value, list):
            for item in value:
                data.append({"parent": key, "label": item})
    return data

# 創建樹狀數據
tree_data = create_tree_data('', topic_tree)

# 將數據轉換為 DataFrame
df = pd.DataFrame(tree_data)

# 使用 Plotly 繪製樹狀圖
fig = px.treemap(df, path=['parent', 'label'], title="互動式樹狀結構")
fig.show()
```
